```python
import pandas as pd
df=pd.read_excel('dataset/SaleData.xlsx')
df
```

```python
df_new=df.groupby('Region').count()[['Sale_amt']]
df_new
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new=df.groupby('Region').sum()[['Sale_amt']]
df_new.reset_index(inplace=True)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
plt.bar(df_new['Region'],df_new['Sale_amt'])
plt.show()
```

```python
df_new=df.groupby('Manager').count()[['Sale_amt']]
df_new
```

```python
df1=df.groupby('Item').sum()['Units']
import matplotlib.pyplot as plt
fig = plt.figure(figsize=[7,5])
ax=fig.add_axes([0,0,1,1])
ax.set_title('items',fontsize=16)
ax.set_ylabel('units',fontsize=12)
ax.set_xlabel('Model year',fontsize=12)
ax.bar(df1.index, df1)
fig.show()
```
